```python
import pyreadline
import imageio
import os, sys
from PIL import Image
from PIL import ExifTags

def CreateGif(Input,Output,Width=1000,Rotate=0,Duration=1,Resize=True):
    filenames = Input
    GIFOutput = Output
    basewidth = Width
    ResizeList = []
    for infile in filenames:
        outfile = os.path.splitext(infile)[0] + "_resize.jpg"
        ResizeList.append(outfile)
        try:
            print('Resizing: ' + infile)
            img = Image.open(infile)
            if Rotate!=0:
                img=img.rotate(Rotate, expand=True)
            print(img.size)
            wpercent = (basewidth/float(img.size[0]))
            hsize = int((float(img.size[1])*float(wpercent)))
            if Resize:
                img = img.resize((basewidth,hsize), Image.ANTIALIAS)
                print(img.size)
            img.save(outfile,quality=100) 
        except IOError:
            print("cannot create thumbnail for " + infile)

    images = []
    for filename in ResizeList:
        images.append(imageio.imread(filename))
    print('Create Gif')
    imageio.mimsave(GIFOutput, images,duration=Duration)
    for filename in ResizeList:
        os.remove(filename)
    print('Done!')
```
